```python
# 1. IMPORT LIBRARIES

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv(r"C:\Users\shnan\OneDrive\Desktop\Fraud.csv")
```

```python
# Drop useless/uninformative columns
drop_cols = ["Unnamed: 0", "trans_num", "trans_date_trans_time", "street", "city", 
             "state", "job", "first", "last"]
df = df.drop(columns=drop_cols, errors="ignore")
```

```python
# 2. HANDLE CATEGORICAL VARIABLES

cat_cols = df.select_dtypes(include=["object"]).columns
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
#3 SPLIT FEATURES / LABEL

X = df.drop("is_fraud", axis=1)
y = df["is_fraud"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
#4. SCALE FEATURES

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#5. TRAIN MODELS

models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Decision Tree": DecisionTreeClassifier(max_depth=8),
    "Random Forest": RandomForestClassifier(n_estimators=200, max_depth=12)
}
```

```python
# 6. EVALUATE MODELS
# ---------------------------------------
for name, model in models.items():
    print("\nLogisticRegression")
    print(f" MODEL: {DecisionTreeClassifier}")
    print("RandomForestClassifier")

    if name == "Logistic Regression":
        model.fit(X_train_scaled, y_train)
        preds = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

    print("Accuracy:", accuracy_score(y_test, preds))
    print(classification_report(y_test, preds))
```
